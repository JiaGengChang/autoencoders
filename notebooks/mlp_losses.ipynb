```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

plt.style.use('ggplot')

SAVEPLOT=True
```

```python
def parsedf(fp):
    df = pd.read_csv(fp, comment='#', names=['iteration','loss'])
    df['iteration'] = [int(_.split()[1]) for _ in df['iteration']]
    df['loss'] = [float(_.split()[1]) for _ in df['loss']]
    return df

def base_plot(ax, df, **kwargs):
    xlab = ax.set_xlabel("Epochs", size=20)
    line = ax.plot(df['iteration'], df['loss'], **kwargs)
    return

def add_plot(ax, df, **kwargs):
    line = ax.plot(df['iteration'], df['loss'], **kwargs)
    
def enlarge_ticklabels(ax):
    xt = ax.get_xticks()
    ax.set_xticklabels(xt, size=20)
    ax.xaxis.set_major_formatter('{x:1.0f}')
    yt = ax.get_yticks()
    ax.set_yticklabels(yt, size=20)
    ax.yaxis.set_major_formatter('{x:0.2f}')
```

```python
df1a = parsedf('output/mlp_lr0.01_m0.csv')
df2a = parsedf('output/mlp_lr0.01_m0.1.csv')
df3a = parsedf('output/mlp_lr0.01_m0.3.csv')
df4a = parsedf('output/mlp_lr0.01_m0.5.csv')
df5a = parsedf('output/mlp_lr0.01_m0.7.csv')
df6a = parsedf('output/mlp_lr0.01_m0.9.csv')

df1b = parsedf('output/mlp_lr0.05_m0.csv')
df2b = parsedf('output/mlp_lr0.05_m0.1.csv')
df3b = parsedf('output/mlp_lr0.05_m0.3.csv')
df4b = parsedf('output/mlp_lr0.05_m0.5.csv')
df5b = parsedf('output/mlp_lr0.05_m0.7.csv')
df6b = parsedf('output/mlp_lr0.05_m0.9.csv')

d1 = {'color': 'blueviolet', 'label':"m = 0.0"}
d2 = {'color': 'darkturquoise', 'label':"m = 0.1"}
d3 = {'color': 'yellowgreen', 'label':"m = 0.3"}
d4 = {'color': 'gold', 'label':"m = 0.5"}
d5 = {'color': 'salmon', 'label':"m = 0.7"}
d6 = {'color': 'firebrick', 'label':"m = 0.9"}
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, sharey=True)
_ = ax1.set_title("Effect of momentum on loss (lr = 0.01)", size=20)
_ = ax2.set_title("Effect of momentum on loss (lr = 0.05)", size=20)
fig.set_figwidth(12)
fig.set_figheight(5)

base_plot(ax1,df1a,**d1)
add_plot(ax1,df2a,**d2)
add_plot(ax1,df3a,**d3)
add_plot(ax1,df4a,**d4)
add_plot(ax1,df5a,**d5)
add_plot(ax1,df6a,**d6)

base_plot(ax2,df1b,**d1)
add_plot(ax2,df2b,**d2)
add_plot(ax2,df3b,**d3)
add_plot(ax2,df4b,**d4)
add_plot(ax2,df5b,**d5)
add_plot(ax2,df6b,**d6)

#newlim = ax.set_xlim(left=-5e2,right=1.1e4)
_ = ax1.legend(prop={'size':20})
_ = ax2.legend(prop={'size':20})

enlarge_ticklabels(ax1)
enlarge_ticklabels(ax2)

plt.tight_layout()

if SAVEPLOT:
    plt.savefig('/local/data/public/2021/jgc47/Desktop/deeplearning/plots/mlp_momentum.png', dpi=300)
```
